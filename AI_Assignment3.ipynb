```python
import pandas as pd
df = pd.read_csv('/content/student-mat_modified.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df.dtypes
```

```python
category_columns = list()
for column in df.columns:
  if df[column].dtype == 'object':
    category_columns.append(column)
```

```python
category_columns
```

```python
mapping_functions = dict()
for column in category_columns:
  values = df[column].unique()
  mapping_function = dict()
  for value_idx, value in enumerate(values):
    mapping_function[value] = value_idx
  mapping_functions[column] = mapping_function
```

```python
mapping_functions
```

```python
for column in category_columns:
  df[column] = df[column].map(mapping_functions[column])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

for column in category_columns:
  df[column] = df[column].astype('category')

X = df.drop('Performance', axis=1)
Y = df['Performance']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model1 = SVC(kernel = 'linear')

model1.fit(X_train, Y_train)

y_pred1 = model1.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred1)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier()

model2.fit(X_train, Y_train)

y_pred2 = model2.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred2)
print(accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB

model3 = GaussianNB()

model3.fit(X_train, Y_train)

y_pred3 = model3.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred3)
print(accuracy)
```
